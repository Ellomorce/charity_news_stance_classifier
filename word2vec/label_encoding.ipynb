```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
data = pd.read_csv('./Original_data/taiwan_charity_news_keywords.csv', encoding='UTF-8')
```

```python
data.head(1)
```

```python
data['Stance_label'].value_counts()
```

```python
Stance_label = data['Stance_label'].tolist()
```

```python
print(Stance_label)
```

```python
# Frequency Encoding
label_freq_map = data['Stance_label'].value_counts().to_dict()
print(label_freq_map)
freq_stance_labels = np.array(list(map(lambda x: label_freq_map.get(x, 0), Stance_label)))
freq_stance_labels = freq_stance_labels.reshape(445,1)
print(freq_stance_labels.shape)
print(freq_stance_labels)
```

```python
#Save freq_stance_label
freq_stance = {'freq_stance_labels': freq_stance_labels}
# Save the dictionary to a compressed '.npz' archive file
np.savez_compressed('./vec_data/freq_stance_labels.npz', **freq_stance)
```

```python
# One-Hot Encoding
ohe = OneHotEncoder()
oh_stance_labels = ohe.fit_transform(data.Stance_label.values.reshape(-1,1)).toarray()
```

```python
print(oh_stance_labels)
```

```python
#Save oh_stance_labels 
oh_stance = {'oh_stance_labels': oh_stance_labels}
# Save the dictionary to a compressed '.npz' archive file
np.savez_compressed('./vec_data/oh_stance_labels.npz', **oh_stance)
```

```python
# Label Encoding
le = LabelEncoder()
le_stance_labels = le.fit_transform(data.Stance_label.values.reshape(-1,1))
```

```python
print(le_stance_labels)
```

```python
#Save le_stance_labels 
le_stance = {'le_stance_labels': le_stance_labels}
# Save the dictionary to a compressed '.npz' archive file
np.savez_compressed('./vec_data/le_stance_labels.npz', **le_stance)
```

```python
def load_npz_file(filepath):
    # Load the numpy array from the .npz file
    with np.load(filepath, allow_pickle=True) as data:
        for key in data.keys():
            arr = data[key]
            break
    return arr
```
