```python
import numpy as np
```

```python
#Load first key in npz
def load_npz_file(filepath):
    # Load the numpy array from the .npz file
    with np.load(filepath, allow_pickle=True) as data:
        for key in data.keys():
            arr = data[key]
            break
    return arr

#Average method: average (30, 300) vectors along with axis 0
def average_vectors(vectors):
    averaged_vectors = []
    # Loop through each vector array in 'vectors', and compute the average vector
    for vector_array in vectors:
        # Compute the mean vector along axis 0
        averaged_vector = np.mean(vector_array, axis=0)
        # Reshape the vector to have shape (1, 300)
        averaged_vector = np.reshape(averaged_vector, (1, -1))
        # Append the averaged vector to the list
        averaged_vectors.append(averaged_vector)
    return averaged_vectors

#Sum method: sum (30, 300) vectors along with axis 0
def sum_vectors(vectors):
    summed_vectors = []
    # Loop through each vector array in 'vectors', and compute the sum vector
    for vector_array in vectors:
        # Compute the summed vector along axis 0
        summed_vector = np.sum(vector_array, axis=0)
        # Reshape the vector to have shape (1, 300)
        summed_vector = np.reshape(summed_vector, (1, -1))
        # Append the averaged vector to the list
        summed_vectors.append(summed_vector)
    return summed_vectors
```

```python
vec20 = load_npz_file(r'D:\Code\NTUT_Thesis\vec_data\vec20.npz')
vec25 = load_npz_file(r'D:\Code\NTUT_Thesis\vec_data\vec25.npz')
vec30 = load_npz_file(r'D:\Code\NTUT_Thesis\vec_data\vec30.npz')
vec35 = load_npz_file(r'D:\Code\NTUT_Thesis\vec_data\vec35.npz')
```
